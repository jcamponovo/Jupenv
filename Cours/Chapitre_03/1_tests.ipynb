Il est utile d'écrire et exécuter des tests lorsque l'on écrit du code. Ces tests servent à vérifier que le code écrit a bien le comportement attendu.  
En python, on peut utiliser la docstring d'une fonction pour écrire ces tests puis le module doctest pour les exécuter. Ainsi, la documentation sera plus complète et contiedra directement des cas d'utilisation de la fonction.  
  

```python
def maxi(liste):
    """
    prend en argument une liste d'entiers
    renvoie l'indice de la première occurence de la valeur maxi
    
    >>> maxi([5,4,9,1])
    2
    
    >>> maxi([5,4,9,1,9,8])
    2
    
    >>> maxi([9,4,1,8])
    0
    
    >>> maxi([1,4,5,9])
    3
    
    >>> maxi([])
    -1
    
    >>> maxi("test")
    
    
    >>> maxi([3.5,5,4])
    
    
    >>> maxi([[1,2,3], [4,5,6]])
    
    
    """
    if len(liste) == 0:
        return -1
    elif type(liste) != list:
        return None
    else:
        imaxi = 0
        for i in range(0,len(liste)):
            if type(liste[i]) != int:
                return None
            if liste[i] > liste[imaxi]:
                imaxi = i
        return imaxi
```

Le TDD ou test driven development est une manière moderne de produire du code de qualité. Le principe est d'écrire dans un premier temps les tests unitaires que la fonction devra passer, puis de coder la fonction qui répond à ces tests de manière la plus simple et rien de plus. On peut ensuite itérer en ajoutant des tests et des fonctionnalités, les tests précédents sont conservés et seront donc testés au fur et à mesure du développement. On est ainsi sûrs que rien n'est cassé lorsque l'on ajoute de nouvelles fonctionalités.  
  
Cette méthode produit du code de qualité car le code est bien testé, et diminue le temps de développement car elle réduit le temps passé à débuguer un programme.
  
Elle oblige aussi à bien se poser la question en terme de spécifications et de comportement attendu pour la fonction écrite avant de commencer à l'écrire.

```python
def tri(liste, reverse = False):
    """
    liste est une liste d'entiers
    Renvoie la liste triée par tri insertion par défaut dans l'ordre croissant
    Reverse est un booléen, si reverse == False, triée dans l'ordre croissant
    si reverse == True, ordre décroissant
    
    >>> tri([1, 3, 2])
    [1, 2, 3]
    
    >>> tri([1, 3, 2], reverse = True)
    [3, 2, 1]
    
    >>> tri([])
    []
    
    >>> tri("test")
    
    
    >>> tri([1,2.5,3])
    
    """
    if type(liste) != list:
        return
    for i in range(len(liste)):
        if type(liste[i]) != int:
            return
        j = i
        while j > 0 and liste[j] < liste[j-1]:
            liste[j], liste[j-1] = liste[j-1], liste[j]
            j = j - 1
    if reverse:
        return liste[::-1]
    else:
        return liste
```
