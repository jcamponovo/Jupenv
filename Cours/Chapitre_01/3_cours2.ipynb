# Cours 2. Entiers relatifs

Les valeurs sont stockées sous forme binaire en mémoire dans l'ordinateur. Comment peut-on alors stocker une valeur négative ?

Une première idée serait d'utiliser le premier bit du code binaire comme bit de signe et coder le reste du binaire comme les valeurs positives.

Ainsi, si $3$ se code sur 8 bits avec $00000011$, avec ce principe, on pourrait coder $-3$ avec $10000011$. 

Ce codage très simple a pour avantage d'être très proche au niveau raisonnement de ce que l'on a l'habitude de manipuler. Cependant, la somme de deux binaires devient alors très difficile à réaliser.

Une deuxième idée est d'utiliser le complément à deux afin de faciliter l'addition entre deux binaires relatifs.  
Pour obtenir le codage d'un nombre négatif, on prend le code du nombre positif correspondant et on inverse tous les bits.

Ainsi, si $3$ se code sur 8 bits avec $00000011$, avec ce principe, on pourrait coder $-3$ avec $11111100$.

Afin d'obtenir un fonctionnement simple lors des additions de binaires relatifs et un codage unique pour le zéro, le système retenu est celui du complément à deux. Ainsi pour obtenir le codage binaire d'un nombre négatif, il faut:  
1. coder le binaire positif correspondant  
2. inverser tous les bits  
3. ajouter 1  
  
**Exemple:**  
$3$ se code sur 8 bits avec $00000011$  
On inverse tous les bits: $11111100$  
Puis on ajoute 1: $11111101$  
Ainsi $11111101$ permet de coder $-3$ en mémoire.  
Si on fait la somme des deux: $00000011 + 11111101 = 1\ 00000000$ la valeur état stocké sur 8 bits, le neuvième bit est perdu et il ne reste en mémoire que $00000000$ soit $0$.  
  
De manière générale, changer le signe se fait en inversant tous les bits du binaire spuis en ajoutant 1.
