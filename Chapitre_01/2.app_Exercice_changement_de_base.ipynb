```python
from ipywidgets import RadioButtons,HBox, VBox, HTML,Text, Button
import random

def on_button_clic(b):
    global val,GUI,sujet,reponse,entree,sortie,boutonRepondre
    if entree.value != sortie.value:
        boutonRepondre.disabled = False
        reponse.value=""
        val = random.randint(0,255)
        if entree.value == 10:
            sujet.value="Convertir le décimal {} en base {}".format(val,sortie.value)
        elif entree.value == 2:
            sujet.value="Convertir le binaire {} en base {}".format(bin(val)[2:].zfill(8),sortie.value)
        else:
            sujet.value="Convertir l'hexadécimal {} en base {}".format(hex(val)[2:].upper(),sortie.value)
    else:
        sujet.value = "Attention les bases choisies sont identiques"

def on_reponse(b):
    global val,GUI,sujet,reponse,entree,sortie,boutonRepondre
    boutonRepondre.disabled = True
    
    if sortie.value == 10:
        if reponse.value == str(val): 
            sujet.value += "</br>Bravo !"
        else:
            sujet.value += "</br>Dommage! La bonne réponse était {}".format(val)
    elif sortie.value == 2:
        if (reponse.value == bin(val)[2:].zfill(8) or reponse.value == bin(val)[2:]):
            sujet.value += "</br>Bravo !"
        else:
            sujet.value += "</br>Dommage! La bonne réponse était {}".format(bin(val)[2:].zfill(8))
    elif sortie.value == 16:
        if (reponse.value == hex(val[2:]) or reponse.value == hex(val[2:]).upper()):
            sujet.value += "</br>Bravo !"
        else:
            sujet.value += "</br>Dommage! La bonne réponse était {}".format(hex(val[2:]).upper())
    
        
entree = RadioButtons(options = [("Base 10",10), ("Base 2",2), ("Base 16",16)])
sortie = RadioButtons(options = [("Base 10",10), ("Base 2",2), ("Base 16",16)])
sortie.value = 2
boutonDemarrer = Button(description = "Démarrer",tooltip="Demarrer")
sujet = HTML("</br>")
reponse = Text()
boutonRepondre = Button(description = "Répondre",tooltip = "Répondre",disabled=True)
GUI = VBox([HTML("<h1><center>changement de base</center></h1>"),HBox([VBox([HTML("<h3>Base de départ</h3>"),entree]),VBox([HTML("<h3>Base d'arrivée</h3>"),sortie])]),
      boutonDemarrer,sujet,HBox([reponse,boutonRepondre])])
boutonDemarrer.on_click(on_button_clic)
boutonRepondre.on_click(on_reponse)
GUI
```
