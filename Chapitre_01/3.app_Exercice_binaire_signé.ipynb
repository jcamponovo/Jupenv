```python
from ipywidgets import RadioButtons,HBox, VBox, HTML,Text, Button
import random

def on_button_clic(b):
    global val,binaire,GUI,sujet,reponse,entree,boutonRepondre
    boutonRepondre.disabled = False
    reponse.value=""
    val = random.randint(-128,127)
    if val < 0:
        binaire = "1"+bin(val+128)[2:].zfill(7)
    else:
        binaire = bin(val)[2:].zfill(8)
    if entree.value == 10:
        sujet.value="Convertir le binaire signé {} codé sur 8 bits en complément à deux en base 10".format(binaire)
    else:
        sujet.value="Convertir l'entier relatif {} en binaire signé codé sur 8 bits en complément à deux".format(val)

def on_value_change(change):
    global val,binaire,GUI,sujet,reponse,entree,boutonRepondre
    boutonRepondre.disabled = True
    sujet.value = "</br>"

    
def on_reponse(b):
    global val,binaire,GUI,sujet,reponse,entree,boutonRepondre
    boutonRepondre.disabled = True
    
    if entree.value == 10:
        if reponse.value == str(val): 
            sujet.value += "</br>Bravo !"
        else:
            sujet.value += "</br>Dommage! La bonne réponse était {}".format(val)
    elif entree.value == 2:
        if reponse.value == binaire:
            sujet.value += "</br>Bravo !"
        else:
            sujet.value += "</br>Dommage! La bonne réponse était {}".format(binaire)
    
        
entree = RadioButtons(options = [("De binaire en base 10",10), ("De base 10 en binaire",2)])
boutonDemarrer = Button(description = "Démarrer",tooltip="Demarrer")
sujet = HTML("</br>")
reponse = Text()
boutonRepondre = Button(description = "Répondre",tooltip = "Répondre",disabled=True)
GUI = VBox([HTML("<h1><center>Entiers relatifs</center></h1>"),HTML("<h3>Sens de conversion</h3>"),entree,
      boutonDemarrer,sujet,HBox([reponse,boutonRepondre])])
boutonDemarrer.on_click(on_button_clic)
boutonRepondre.on_click(on_reponse)
entree.observe(on_value_change, names='value')
GUI
```
