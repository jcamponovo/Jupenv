```python
from ipywidgets import FileUpload, Button, HBox, Layout, Dropdown, VBox,Label, Box
from ipywidgets import HTML as html
from IPython.core.display import HTML
from IPython.display import Javascript
import os
liste_f = ["---"] # + [val for val in os.listdir() if len(val)>6 and val[-6:]==".ipynb" and val!="new.ipynb" and val[0]!="."]
liste_f.sort()
lock_bou3 = False
new = 0
liste_e = ["---"] + ["formation","cours"] + [val for val in os.listdir("./personnalisation_2020/") if len(val)==2 and val[0]!="."]
liste_e.sort()
base = {}
for val in liste_f:
    base[val] = ""

bou = Button(
    description='Nouveau Notebook',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Nouveau Notebook',
    icon='' # (FontAwesome names without the `fa-` prefix)
)

bou5 = Button(
    description='Nouveau .py',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Nouveau .py',
    icon='' # (FontAwesome names without the `fa-` prefix)
)

bou4 = Dropdown(
    options=liste_e,
    value="---",
    description='Catégorie',
    disabled=False,
)

bou3 = Dropdown(
    options=liste_f,
    value="---",
    description='Ouvrir',
    disabled=False,
)

f = FileUpload(
    description = "Charger",
    accept='.ipynb',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
    multiple=False  # True to accept multiple files upload else False
)


bou2 = Button(
    description='Terminal',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Terminal',
    icon='' # (FontAwesome names without the `fa-` prefix)
)

def change(b):
#    (os.path.join(root, i),
    global lock_bou3,base
    if bou3.value !="---":
        display(Javascript('window.open("../notebooks/'+base[bou3.value]+bou3.value+'");'))
    #else:
    #    lock_bou3 = False


def change2(b):
    global lock_bou3,liste_f,bou3,bou4,base
    #lock_bou3 = True
    liste_f = ["---"]
    if bou4.value == "cours":
        i = 0
        while os.path.exists("./Cours_{:02d}".format(i)):
            liste2 = [val for val in os.listdir("./Cours_{:02d}".format(i)) if len(val)>6 and val[-6:]==".ipynb" and val!="new.ipynb" and val[0]!="."]
            for val in liste2:
                base[val] = "./Cours_{:02d}/".format(i)
            i += 1
            liste_f += liste2[:]
        
    elif bou4.value == "formation":
        liste_f += [val for val in os.listdir() if len(val)>6 and val[-6:]==".ipynb" and val!="new.ipynb" and val[0]!="."]
        for val in liste_f:
            base[val] = ""
    elif bou4.value != "---":
        liste_f += [val for val in os.listdir("./personnalisation_2020/{}".format(bou4.value)) if len(val)>6 and val[-6:]==".ipynb" and val!="new.ipynb" and val[0]!="."]
        for val in liste_f:
            base[val] = "personnalisation_2020/{}/".format(bou4.value)
    liste_f.sort()
    bou3.options = liste_f
    bou3.value = liste_f[0]

        
    
def on_use(b):
    v = f.value
    if len(v)>6 and v in os.listdir():
        v = v[:-6]+(1)+v[-6:]
    with open(next(iter(f.value)), "w+b") as i:
        i.write(f.data[0])
    os.popen("jupyter trust '{}'".format(next(iter(f.value))))
    url = "../notebooks/"+ next(iter(f.value))
    display(Javascript('window.open("{url}");'.format(url=url)))
    #display(HTML('<a href="../notebooks/note.ipynb">go !</a>'))

def on_click2(b):
    display(Javascript('window.open("../terminals/1");'))
    #display(HTML('<a href="../terminals/1">go !</a>'))
    
def on_click(b):
    global new
    new += 1
    i = open("new.ipynb","r")
    j = open("new{}.ipynb".format(new),"w")
    j.write(i.read())
    i.close()
    j.close()
    url = "../notebooks/new{}.ipynb".format(new)
    display(Javascript('window.open("{url}");'.format(url=url)))
    # display(HTML('<a href="../notebooks/new.ipynb">go !</a>'))
    
def on_click3(b):
    global new
    new += 1
    j = open("new{}.py".format(new),"w")
    j.close()
    url = "../edit/new{}.py".format(new)
    display(Javascript('window.open("{url}");'.format(url=url)))
    # display(HTML('<a href="../notebooks/new.ipynb">go !</a>'))
    
bou.on_click(on_click)
bou2.on_click(on_click2)
bou5.on_click(on_click3)
bou3.observe(change,"value")
bou4.observe(change2,"value")
f.observe(on_use,"data")
display(HTML("""<h1 style="text-align:center; width:100%;"> Environnement JUPYTER </h1>"""))
display(Box([VBox([html("""<h4 style="text-align:center; width:100%;">Ouvrir un Cours/TD</h4>"""),bou4,bou3],layout=Layout(min_width="350px",flex_flow="column wrap")),VBox([html("""<h4 style="text-align:center; width:100%;">Créer/Charger un notebook</h4>"""),bou,bou5,f],layout=Layout(min_width="200px",flex_flow="column wrap")),VBox([html("""<h4 style="text-align:center; width:100%;">Terminal Linux</h4>"""),bou2],layout=Layout(flex_flow="column wrap",min_width="200px"))],layout=Layout(width="100%",height="auto",display="flex",flex_flow="row wrap",min_width='800px',justify_content='space-around')))
```
