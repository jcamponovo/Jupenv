```python
import cv2
import matplotlib.pyplot as plt
```

Les pixels sont écrits de gauche à droite sur la ligne la plus en haut de l'image puis de gauche à droite sur la deuxième ligne etc... jusqu'à avoir balayé toute l'image. On ne travail pas en coordonnées des pixels mais on écrit les pixels dans cet ordre les uns à la suite des autres

```python
f = open("ger.ppm", "w")
f.write("P3\n")
f.write("300 180\n")
f.write("255\n")

# 60 lignes identiques de 300 pixels noirs
for i in range(60):
    for j in range(300):
        f.write("0 0 0\n")
        
# 60 lignes identiques de 300 pixels rouges
for i in range(60):
    for j in range(300):
        f.write("255 0 0\n")
        
# 60 lignes identiques de 300 pixels jaunes
for i in range(60):
    for j in range(300):
        f.write("255 255 0\n")
f.close()
```

```python
img = cv2.imread("ger.ppm")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.show()
```

```python
f = open("greece.ppm", "w")
f.write("P3\n")
f.write("300 180\n")
f.write("255\n")

for i in range(20): # 20 lignes identiques debut de la croix et bande bleue
    f.write("30 30 255\n"*40+"255 255 255\n"*20+"30 30 255\n"*240)
for i in range(20): # 20 lignes identiques suite d ela croix et bande blanche
    f.write("30 30 255\n"*40+"255 255 255\n"*20+"30 30 255\n"*40+"255 255 255\n"*200)
for i in range(20): # 20 lignes identiques milieu de la croix et bande bleue 
    f.write("255 255 255\n"*100+"30 30 255\n"*200)
for i in range(20): # 20 lignes identiques suite de la croix et bande blanche
    f.write("30 30 255\n"*40+"255 255 255\n"*20+"30 30 255\n"*40+"255 255 255\n"*200)
for i in range(20): # 20 lignes identiques fin de la croix et bande bleue
    f.write("30 30 255\n"*40+"255 255 255\n"*20+"30 30 255\n"*240)
for i in range(20): # 20 lignes identiques bande blanche
    f.write("255 255 255\n"*300)
for i in range(20): # 20 lignes identiques bande bleue
    f.write("30 30 255\n"*300)
for i in range(20): # 20 lignes identiques bande blanche
    f.write("255 255 255\n"*300)
for i in range(20): # 20 lignes identiques bande bleue
    f.write("30 30 255\n"*300)
f.close()
```

```python
f = open("scotland.ppm", "w")
f.write("P3\n")
f.write("300 180\n")
f.write("255\n")

blue = "30 30 255\n"
white = "255 255 255\n"

# on balaye tout le drapeau l'ordre de x et y a son importance car on doit les écrire dans l'ordre
for y in range(180):
    for x in range(300): # deux bandes d'équation y = +180/300 * x et y = -180/300 * x + 180
                         # on utilise ici une largeur de 20 pixels
        if x*180/300 - 20 <= y <= x*180/300 + 20 or - x*180/300 + 180 - 20<= y <= - x*180/300 + 180 + 20:
            f.write(white)
        else:
            f.write(blue)
f.close()
```

```python
img = cv2.imread("scotland.ppm")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.show()
```

```python
f = open("japan.ppm", "w")
f.write("P3\n")
f.write("300 180\n")
f.write("255\n")
red = "255 0 0\n"
white = "255 255 255\n"
rayon = 45
xcentre = 150
ycentre = 90

# On balaye toute l'image
for y in range(180):
    for x in range(300):
        if (x-xcentre)**2 + (y-ycentre)**2 <= rayon**2: # si la distance au carré entre le centre et le pixel
                                                        # est inférieure au rayon du cercle au carré,
            f.write(red) # alors le pixel est rouge
        else:
            f.write(white) # sinon il est blanc
f.close()
```

```python
img = cv2.imread("japan.ppm")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.show()
```

```python
# travail en coordonnées pour ceux qui savent déjà manipuler les listes

f = open("fr3.ppm", "w")
f.write("P3\n")
f.write("300 180\n")
f.write("255\n")
tab = [["255 255 255\n" for j in range(300)] for i in range(180)] # fond tout blanc

for x in range(0,100): # rectangle rouge
    for y in range(0,180):
        tab[y][x] = "255 0 0\n"
        
for x in range(200,300): # rectangle bleu
    for y in range(0,180):
        tab[y][x] = "0 0 255\n"
        
# ecriture des pixels dans le fichier
for ligne in tab:
    f.write("".join(ligne))
f.close()
```

```python
img = cv2.imread("fr3.ppm")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.show()
```

```python
f = open("grece2.ppm", "w")
f.write("P3\n")
f.write("300 180\n")
f.write("255\n")
tab = [["255 255 255\n" for j in range(300)] for i in range(180)] # fond tout blanc
bleu = "30 30 255\n"

for xc in range(0,100,60): # 4 carrés bleus de 40x40 pour dessiner la croix
    for yc in range(0,100,60):
        for x in range(40):
            for y in range(40):
                tab[y+yc][x+xc] = bleu
                
for yc in range(0,100,40): # bandes bleues à droite du carré
    for x in range(100,300):
        for y in range(0,20):
            tab[y+yc][x] = bleu
            
for yc in range(120,180,40): # bande bleues en dessous du carré
    for x in range(0,300):
        for y in range(0,20):
            tab[y+yc][x] = bleu
for ligne in tab:
    f.write("".join(ligne))
f.close()
```

```python
img = cv2.imread("grece2.ppm")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.show()
```

```python
f = open("grece3.ppm", "w")
f.write("P3\n")
f.write("300 180\n")
f.write("255\n")

bleu = "30 30 255\n" # code pixel bleu
blanc = "255 255 255\n" # code pixel blanc
tab = [[blanc for j in range(300)] for i in range(180)] # tout blanc

# bandes bleues partout
for yc in range(0,180,40):
    for x in range(0,300):
        for y in range(0,20):
            tab[y+yc][x] = bleu

# carré bleu de 100x100 en haut à gauche par dessus les bandes
for y in range(0,100):
    for x in range(0,100):
        tab[y][x] = bleu 

# croix blanche en haut à gauche par dessus le carré bleu
for y in range(40,60): 
    for x in range(0,100):
        tab[y][x] = blanc 
for y in range(0,100):
    for x in range(40,60):
        tab[y][x] = blanc

# ecriture des pixels dans le fichier
for ligne in tab:
    f.write("".join(ligne))
f.close()
```

```python
img = cv2.imread("grece3.ppm")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.show()
```

```python
# Exemple avec le tracé d'une cardioide

from math import cos,sin,pi

f = open("cardio.ppm", "w")
f.write("P3\n")
f.write("300 300\n")
f.write("255\n")

pink = "255 0 255\n"
white = "255 255 255\n"
rayon = 50
tab = [[blanc for j in range(6*rayon)] for i in range(6*rayon)] # tout blanc

xcentre = 3*rayon
ycentre = rayon

# On balaye toute l'image
for param in range(1001):
    phi = 2*pi*param/1000
    yc = int(ycentre-2*rayon*(1-cos(phi))*cos(phi)) # equation de la courbe
    xc = int(xcentre+2*rayon*(1-cos(phi))*sin(phi))
    for x in range(xc-5,xc+5): # brosse carrée de 5x5
        for y in range(yc-5,yc+5):
            tab[y][x] = pink
# ecriture des pixels dans le fichier
for ligne in tab:
    f.write("".join(ligne))
    
f.close()
```

```python
img = cv2.imread("cardio.ppm")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.show()
```
